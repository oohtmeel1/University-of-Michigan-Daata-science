## Set up the environment

### Task: import packages and rename them accordingly

````python
import seaborn as sns
df = sns.load_dataset('Iris')
````

## Load the dataset

````python
df
````

## Play with relational plots

````python
sns.relplot(data = df, x="sepal_length", y="sepal_width", hue="species")
````

## Play with distribution plots

````python
iris = sns.load_dataset("iris")
sns.displot(iris, x="sepal_length")
````

## Play with other plots

````python
import numpy as np
sns.catplot(data=iris, x="species", y="sepal_length", hue="species", kind="swarm")
````
