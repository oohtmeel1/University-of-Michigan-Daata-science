## Set up the environment

### Task: import packages and rename them accordingly

````python
import seaborn as sns
import pandas as pd
````

## Load the dataset

````python
df = sns.load_dataset('iris')
df
````

## Play with distribution plots
You can show the data distribution for some attributes interest you.

````python
sns.scatterplot(data = df, x = "petal_length", y = "species")
````

## Play with relational plots
You can show how two or more attributes are associated.

````python
sns.relplot(data = df, x = "petal_length", y = "petal_width", hue ="species")
````

## Play with other plots
You can play with other plots we learned here. Be as creative as you can and see how the visualization helps you to understand the data better.

````python
import seaborn as sns
sns.set_theme()

df = sns.load_dataset("iris")
sns.catplot(
    data=df, x="petal_length", y="petal_width", hue="species",
    native_scale=True, zorder=1
)
sns.regplot(
    data=df, x="petal_length", y="sepal_width",
    scatter=True, truncate=False, order=10, color=".5",
)
````
